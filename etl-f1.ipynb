## Librerias 

```python
import  requests
import pandas as pd
```

```python
url='https://ergast.com/api/f1/2022/22/results.json'
```

```python
r=requests.get(url)
json=r.json()
```

Veremos que claves contiene nuestro json

```python
json.keys()
```

```python
json['MRData'].keys()
```

Podemos observar que solo tiene una clave. Ahora nosotrps queremos ver las carreras. 

```python
json['MRData']['RaceTable']['Races'][0]
```

Esto ya es un diccionario

```python
json['MRData']['RaceTable']['Races'][0].keys()
```

```python
Ahora trabajaremos con los resultados de las carreras
```

```python
results=pd.DataFrame(json['MRData']['RaceTable']['Races'][0]['Results'])
```

```python
results.head()
```

```python
results['Driver']
```

A cada una de las entradas de drive, le vamos a aplicar la funcion ... para extraer  cada elemento de la columna driver

```python
results['Driver']=results['Driver'].apply(lambda x:x['driverId'])
```

```python
results.head()
```

```python
results['Constructor']
```

```python
results['Constructor']=results['Constructor'].apply(lambda x:x['constructorId'])
```

```python
results['FastestLap']=results['FastestLap'].apply(lambda x:x['rank'])
```
