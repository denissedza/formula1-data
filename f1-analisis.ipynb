Agregamos primero las librerias con las que vamos a trabajar.

```python
import requests
import json
import pandas as pd
#warnings.filterwarning('ignore') Hacemos esto para omitir los warnings 
```

Asignamos a la variable url la url de la api donde estan nuestros resultados de la carrera de 2022

```python
url='https://ergast.com/api/f1/2022/22/results.json'
```

Despues asignamos a la variable response, le pedimos que obtenga los datos que le estamos pidiendo de la url

```python
response=requests.get(url)
```

Con el atributo content podemos observar lo que contiene response

```python
r=response.content
```

Pero para que no se vea asi le damos el formato de json para que quede como diccionario

```python
content=json.loads(response.content)
```

Haremos un for para tener las carreras del 2022. Lo que tiene que cambiar es el url donde pide.
Nota.
Ponemos {} para señalar que es una variable aleaotoria y va cambiando.

```python
df_results_2022=pd.DataFrame()
x=pd.DataFrame()
for race in  list(range(1,23)): 
    url='https://ergast.com/api/f1/2022/{}/results.json'
    response=requests.get(url. format(race))# Aplicamos la funcion format que indica que es lo que va cambiando lo que es race
    #Al final solo amos a ver el ultimo elemento que  se consulta
    content=json.loads(response.content)
    #Por eso vamos agregando data frame con data frame.
    #df_results_2022=df_results_2022.append(pd.DataFrame(content['MRData']['RaceTable']['Races'][0]['Results']))
    
    df_results_2022=pd.concat([df_results_2022,pd.DataFrame(content['MRData']['RaceTable']['Races'][0]['Results']))
    
```

```python
 df_results_2022['Driver']= df_results_2022['Driver'].apply(lambda x:x['driverId'])
```

Ahora queremos sumar los puntos de cada piloto para esto usaremos el metodo groupby de pandas.

```python
df_results_2022.groupby(['position'])
```

```python
Convertimos los puntos de la columna points en datos numericos
```

```python
df_results_2022['points']=df_results_2022.points.astype(int)
```
