##  Tabla Drivers

Primero haremos nuestro diccionario de listas, llamado drivers.

```python
drivers={'year':[],
        'id': [],
        'permanentNumber':[],
        'code':[],
        'url':[],
        'givenName':[],
        'familyName':[],
        'dob':[],
        'nationality':[],
        }
```

Haremos un for para tener las carreras del 2022. Lo que tiene que cambiar es el url donde piden los años
NOTA:
Ponemos {} para señalar que es una variable aleatoria y va cambiando.

```python
for year in list(range(2010,2023)):    
    url = 'http://ergast.com/api/f1/{}/drivers.json'
    response = requests.get(url.format(year))
    content=json.loads(response.content)

    for item in content['MRData']['DriverTable']['Drivers']:
        drivers['year'].append(year)
        drivers['id'].append(item['driverId'])
        drivers['code'].append(item['code'])
        drivers['url'].append(item['url'])
        drivers['givenName'].append(item['givenName'])
        drivers['familyName'].append(item['familyName'])
        drivers['dob'].append(item['dateOfBirth'])
        drivers['nationality'].append(item['nationality'])
        try:
            drivers['permanentNumber'].append(item['permanentNumber'])
        except:
            drivers['permanentNumber'].append(None)
```

## Data frame Drivers

```python
pd.DataFrame(drivers).to_csv('drivers.csv',index=False)
```
